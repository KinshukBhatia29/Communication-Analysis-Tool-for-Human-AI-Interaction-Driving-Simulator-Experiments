```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_histogram(df):
    """Plots a histogram of transcribed words per 5-second time bucket."""
    bucket_counts = df.groupby("bucket_start").size()
    
    plt.figure(figsize=(10, 5))
    plt.bar(bucket_counts.index, bucket_counts.values, width=4, color='skyblue')
    plt.xlabel("Time (seconds)")
    plt.ylabel("Number of Words")
    plt.title("Word Count Distribution Over Time")
    plt.show()

# Load the CSV file
df_transcripts = pd.read_csv("/kaggle/input/transcription/transcription_results.csv")

# Call the function with the DataFrame
plot_histogram(df_transcripts)
```

```python
import seaborn as sns

def plot_sentiment_distribution(df):
    """Plots sentiment classification results."""
    sentiment_counts = df["sentiment"].value_counts()
    
    plt.figure(figsize=(6, 4))
    sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="coolwarm")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.title("Sentiment Analysis Results")
    plt.show()

# Example usage
plot_sentiment_distribution(df_transcripts)
```
